```python
import numpy as np
from matplotlib import pyplot as plt
import random
import math
import copy

x1 = [-1,-1,1,1]
x2 = [-1,1,-1,1]

#  expected output
t = [1,1,1,0]

# Generating random weights 
w1 = random.uniform(-2.0,2)
w2 = random.uniform(-2.0,2)
# bias 
w3 = random.uniform(-2.0,2)

#Learning rate
lr = 0.1

output = []

flag = False
count = 1
k = np.linspace(-3,3,num= 5)

plt.xlabel("x1-axis")
plt.ylabel("x2-axis")
plt.axis([-1.5,3,-1.5,3])

for i in range(0,4):
    if t[i]>0:
        # mark output one using a blue circle
        plt.scatter(x1[i],x2[i],marker="o",c="b")
    else:
        # mark output zero using a red square
        plt.scatter(x1[i],x2[i],marker="s",c="r")

l = -((w3+(k*w1))/w2)
plt.plot(k,l,linestyle='--',c='r')

plt.show()
```

```python
plt.xlabel("x1-axis")
plt.ylabel("x2-axis")
plt.axis([-1.5,3,-1.5,3])

for i in range(0,4):
    if t[i]>0:
        plt.scatter(x1[i],x2[i],marker="o",c="b")
    else:
        plt.scatter(x1[i],x2[i],marker="s",c="r")

l = -((w3+(k*w1))/w2)
plt.plot(k,l,linestyle='-',c='r')
plt.plot(k,l,linestyle='-',c='r')
while count < 10000:
    output = []
    for i in range(0,4):
        
        u = (x1[i]*w1) + (x2[i]*w2) + w3
        
        y = 1.0/( 1 + np.exp(-u))

        dw1 = lr*(x1[i]*(t[i]-y))*(y*(1-y))
        dw2 = lr*(x2[i]*(t[i]-y))*(y*(1-y))
        dw3 = lr*(t[i]-y)*(y*(1-y))

        w1 = w1 + dw1
        w2 = w2 + dw2
        w3 = w3 + dw3

        output.insert(i,y)

        
    l = -((w3+(k*w1))/w2)

    plt.plot(k,l,linestyle='--',c='g')
        
    
    count = count + 1 

print("Total training steps: ", count-1)
print ("Target =", t)
print ("Actual Output y =", output)
plt.plot(k,l,linestyle='-', c='b')
plt.show()
```
